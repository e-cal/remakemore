```python
import torch
import matplotlib.pyplot as plt

%matplotlib inline
```

# Data

```python
words = open("names.txt", "r").read().splitlines()
```

```python
words[:10]
```

```python
print(f"num words: {len(words)}")
print(f"shortest: {min(len(w) for w in words)}")
print(f"longest: {max(len(w) for w in words)}")
```

```python
chars = [chr(i) for i in range(97, 97 + 25 + 1)]  # 'a' ... 'z'
stoi = {s: i + 1 for i, s in enumerate(chars)}
stoi["."] = 0  # start/end indicator
itos = {i: s for s, i in stoi.items()}
```

```python
# count bigrams (ordered letter pairs in dataset)

b = torch.zeros((27, 27), dtype=torch.int32)

for word in words:
    chrs = ["."] + list(word) + ["."]
    for c1, c2 in zip(chrs, chrs[1:]):
        # bigram = (c1, c2)
        b[stoi[c1], stoi[c2]] += 1
```

```python
# visualize

plt.figure(figsize=(16, 16))
plt.imshow(b, cmap="YlOrRd")
plt.axis("off")
for i in range(27):
    for j in range(27):
        chstr = itos[i] + itos[j]
        plt.text(j, i, b[i, j].item(), ha="center", va="top", color="gray")
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")
```
